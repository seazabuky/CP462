# 6
**filter data before merging**

```python
left = movies[movies['movie title'].str[-5:-1] > '1990'].iloc[:, 0:3]
left
```

**merging dataframe with left 'movie id' and right 'item_id' columns**

```python
right = ratings.copy()
result = pd.merge(left,right,left_on="movie id",right_on="item_id")
result
```

**Result Top 20 (from tail) Ratings of the Movie **

```python
result = result.groupby(['movie title'])['rating'].mean().sort_values(ascending=False).tail(20)
result
```
